```python
import yfinance as yf
import pandas as pd
```

```python
msft = yf.Ticker("MSFT")

# get all stock info
msft.info
```

```python
# get historical market data
hist = msft.history(period="1y")
print(hist)
```

```python
pd.DataFrame(hist).to_csv('msft_1y.csv')
```

```python
pd.DataFrame(hist).describe()
```

```python
# show meta information about the history (requires history() to be called first)
msft.history()
msft.history_metadata
```
